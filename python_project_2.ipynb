```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset (1).csv")
data.head()
```

```python
type(data)
age_series = data['Age']
fare_series = data['Fare']
survived_series = data['Survived']

print(type(age_series))
```

```python
data.head()
data.info()
data.describe()
```

```python
data["Age"] = data["Age"].fillna(data["Age"].mean())
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
```

```python
data.iloc[0]
data.iloc[0][['Age', 'Fare']]
data.iloc[0:6, 2:6]
```

```python
data.rename(columns={'Sex': 'Gender'}, inplace=True)
pivot_table = data.pivot_table(values='Survived',
                             index='Gender',
                             aggfunc='mean')
melted = pd.melt(data, id_vars=['PassengerId'],
                 value_vars=['Age', 'Fare'])
```

```python

data[data['Age'] > 30]
data[data['Fare'] > 50]
data[(data['Gender'] == 'female') & (data['Survived'] == 1)]
data.sort_values(by='Fare', ascending=False)
data.sort_values(by='Age')
data.sort_values(by='Survived', ascending=False)
```

```python
data.groupby('Gender')['Survived'].mean()
data.groupby('Pclass')['Age'].mean()
data.groupby('Embarked')['Survived'].sum()
```

```python
class_info = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'Category': ['Upper', 'Middle', 'Lower']
})
merged_df = pd.merge(data, class_info, on='Pclass')
data.join(class_info.set_index('Pclass'), on='Pclass')
concat_data = pd.concat([data.head(), data.tail()])
```
